```python
import pandas as pd 
my_dict={'sale_id':[1,2,3,4,5,6,7,8,9],
         'c_id':[2,2,1,4,2,3,2,3,2],
         'p_id':[3,4,3,2,3,3,2,2,3],
         'price':[30,20,40,20,30,40,20,30,40],
         'product':['Monitor','CPU','Monitor','RAM','Monitor','Monitor','RAM','RAM','Monitor'],
         'qty':[2,1,3,2,3,2,3,2,2],
         'store':['ABC','DEF','ABC','DEF','ABC','DEF','ABC','DEF','ABC']}
sales = pd.DataFrame(my_dict)
sales
```

```python
my_sale_d_col=sales.groupby(['product','store'],as_index=False).agg({'qty':'sum','price':'mean'})
my_sale_d_col
```

```python
my_sale_d_col1=sales.groupby(['product','store']).agg({'qty':'sum','price':'mean'})
my_sale_d_col1
```

```python
d={'qty':['sum'],'price':['mean','max','min']}
my_sale_d_col2=sales.groupby(['product','store'],as_index=False).agg(d)
my_sale_d_col2
```
